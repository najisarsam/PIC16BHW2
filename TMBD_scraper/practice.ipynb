```python
import pandas as pd
```

```python
results = pd.read_csv("results.csv")
results = results[~results.duplicated()]
results
```

```python
occurences = pd.DataFrame(results["movie_or_TV_name"].value_counts()).reset_index()
occurences = occurences.rename(columns={"index" : "Name of Film / Show", "movie_or_TV_name" : "Number of Shared Actors With The Blacklist"})
occurences
```

```python
recommendations = occurences[occurences["Name of Film / Show"] != "The Blacklist"]
recommendations.head(10)
```

```python
from plotly import express as px
```

```python
fig = px.histogram(recommendations,
                   x = "Number of Shared Actors With The Blacklist",
                   histnorm = 'percent')
fig.show()
```
